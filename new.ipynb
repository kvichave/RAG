Prompts and template

```python
from langchain_core.prompts import ChatPromptTemplate
prompt= ChatPromptTemplate.from_messages(
    [
        ('system',"you are VENOM, adapt the tone and behaviour of VENOM"),
        ('human','hey, who are you?')
    ]
)


```

CHAPTER 3 â€” Chains

```python
print(llm.invoke(prompt.format(topic="Batman")).content)
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser() 


chain = (
    ChatPromptTemplate.from_template("Write a short note about {topic}.")
    | llm
    | parser
)


res=chain.invoke({"topic":"superman"})

```

```python
print(res)
```
